```python
colors = 'blue green red yellow black'.split(' ')
i = 0
for col in data.drop(columns='Total_score,Average'.split(',')):
    if data[col].dtype !='object':
        sns.kdeplot(data=data,x=data[col],shade=True,color=colors[i],label=col)
        plt.xlabel('scores')
        plt.legend(loc='upper left')
        i+=1
```

```python
plt.figure(figsize=(20,7))
plt.subplot(141)
sns.histplot(data,x='Average',kde=True,bins=30)
plt.subplot(142)
sns.histplot(data,x='Average',kde=True,bins=30,hue='race/ethnicity')
plt.subplot(143)
sns.histplot(data,x='Average',kde=True,bins=30,hue='gender')
```

```python
female_data = data[data['gender']=='female']
male_data = data[data['gender']=='male']
sns.kdeplot(data=female_data,x='Average',shade=True,label='Female')
sns.kdeplot(data=male_data,x='Average',shade=True,label='male')
plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(20,7))
plt.subplot(141)
sns.histplot(data,x='Average',kde=True,bins=30)
plt.subplot(142)
sns.histplot(female_data,x='Average',kde=True,bins=30,hue='lunch')
plt.subplot(143)
sns.histplot(male_data,x='Average',kde=True,bins=30,hue='gender')
```

```python
for val in data['parental level of education'].unique():
    df = data[data['parental level of education']==f"{val}"]
    sns.kdeplot(data=df,x=df['Average'],shade=True,label=val)
plt.legend(loc='upper left')
```
